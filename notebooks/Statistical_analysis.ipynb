Churn Significance Test (Proportion Z-Test)

Churn is binary, so we use a Z-test.

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

df = pd.read_csv("/Users/rahulpoojith/Documents/CRMProject/Datasets/ab_test_data.csv")

stats = df.groupby("experiment_group")["churned"].agg(["sum", "count"])
print(stats)

count = stats["sum"].values
nobs  = stats["count"].values

z_stat, p_value = proportions_ztest(count, nobs)

z_stat, p_value
print(f"z_stat: {z_stat}, p_value: {p_value}")
```
